```python
from io import open
import pandas as pd
import math

pathsToTexts = [r'C:\Users\mrrai\PycharmProjects\pythonProject6\task_1\text_1.txt', r'C:\Users\mrrai\PycharmProjects\pythonProject6\task_1\text_2.txt', r'C:\Users\mrrai\PycharmProjects\pythonProject6\task_1\text_3.txt', r'C:\Users\mrrai\PycharmProjects\pythonProject6\task_1\text_4.txt', r'C:\Users\mrrai\PycharmProjects\pythonProject6\task_1\text_5.txt', r'C:\Users\mrrai\PycharmProjects\pythonProject6\task_1\text_6.txt', r'C:\Users\mrrai\PycharmProjects\pythonProject6\task_1\text_7.txt', r'C:\Users\mrrai\PycharmProjects\pythonProject6\task_1\text_8.txt', r'C:\Users\mrrai\PycharmProjects\pythonProject6\task_1\text_9.txt', r'C:\Users\mrrai\PycharmProjects\pythonProject6\task_1\text_10.txt']
numberOfTexts = len(pathsToTexts)

# retrieving texts into a list
textsList = []
for i in range(numberOfTexts):
    textsList.append(open(pathsToTexts[i], encoding='utf-8').read())

# filtering out dots, etc.
for i in range(numberOfTexts):
    textsList[i] = textsList[i].replace('.', '').replace(',', '').replace('?', '').replace('!', '').replace(':', '').replace('(', '').replace(')', '').replace("'s", '').replace('-', '').replace('‘', '').replace('#', '').replace("'", '').lower()




# converting into lists of lists-of-words
wordlists = []
for text in textsList:
    wordlists.append(text.split())

# forming dictionary
totalDict = dict()
for i in range(numberOfTexts):
    for word in wordlists[i]:
        if word in totalDict:
            totalDict[word][i] += 1
        else:
            totalDict[word] = [0] * 10
print(totalDict)


# creating a set of all words encompassing all texts
allWordsSet = set(wordlists[0])
for i in range(numberOfTexts):
    for word in wordlists[i]:
        allWordsSet.add(word)
    #allWordsSet.union(set(wordlists[i]))

# creating dictionaries for each text to keep occurrence of each word in them
numberOfWordsInTextList = []
for i in range(numberOfTexts):
    numberOfWordsInTextList.append(dict.fromkeys(allWordsSet, 0))
    for word in wordlists[i]:
            numberOfWordsInTextList[i][word] += 1

# calculating tf value for every word in every document
tfValueList = []
for i in range(numberOfTexts):
    dict = {}
    for word, count in numberOfWordsInTextList[i].items():
        dict[word] = count/float(len(wordlists[i]))
    tfValueList.append(dict)

# calculating ifd
idfDict = dict.fromkeys(numberOfWordsInTextList[0].keys(), 0)
for numberOfWordsInText in numberOfWordsInTextList:
    for word, val in numberOfWordsInText.items():
        if val > 0:
            idfDict[word] += 1
for word, val in idfDict.items():
    idfDict[word] = math.log(numberOfTexts / float(val))


# calculating tfidf
tfidfVectorValuesList = []
tfidf = {}
for tfValue in tfValueList:
    for word, value in tfValue.items():
        tfidf[word] = value * idfDict[word]
    tfidfVectorValuesList.append(tfidf)
    tfidf = {}

# displaying vectors
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df = pd.DataFrame(tfidfVectorValuesList)
print(df)
```
